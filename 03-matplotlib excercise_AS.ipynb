```python
"** Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**"
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,.4,.4])

axes1.set_title('Main PLot')
axes2.set_title('Inner plot')

axes1.plot(x,z)
axes2.plot(x,y)
```

```python
"** Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**"
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax1.plot(x,z, label='x**2', color='red')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('Main plot x VS z')

ax2 = fig.add_axes([0.2,0.5,.4,.4])
ax2.plot(x,y,label='x*2',color='blue')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('Inner plot x VS y')

plt.show()
```

```python
"** Use plt.subplots(nrows=1, ncols=2) to create the plot below.**"
```

```python
import matplotlib.pyplot as plt

fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10,4))

ax1.plot(x,y, label='x*y',color='red')
ax1.set_title('Plot 1: x VS y')
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2.plot(x,z, label='x**2',color='blue')
ax2.set_title('Plot 2: x VS z')
ax2.set_xlabel('x')
ax2.set_ylabel('y')

plt.show()
```

```python
"** Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**"
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8,5))

ax.plot(x,y, color='red',linewidth=2.5, linestyle='--', label='x*2')
ax.plot(x,z, color='blue',linewidth=1.5,linestyle=':',label='x**2')

ax.set_title('Stylized plot of x VS y and x VS z')
ax.set_xlabel('x')
ax.set_ylabel('values')

plt.show()
```
